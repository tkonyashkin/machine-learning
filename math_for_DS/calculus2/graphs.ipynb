```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 50)
y = np.linspace(0, 10, 50)
x, y = np.meshgrid(x, y)

z = np.sin(x*y)

plt.contour(x, y, z, levels=[0])

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z)

plt.show()
```
