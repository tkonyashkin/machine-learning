```python
df = pd.read_sql("""SELECT  f.post_id,
                            f.time,
                            u.age,
                            u.os
                    FROM "feed_action" f
                    INNER JOIN "user" u
                    ON f.user_id = u.id
                    WHERE u.city = 'Omsk'
                    AND f.action = 'like'
                    ORDER BY f.time DESC
                    LIMIT 100""", con)
```

```python
df = pd.read_sql("""SELECT  u.city,
                            u.age,
                            DATE(f.time),
                            p.topic,
                            f.action
                    FROM "feed_action" f
                    INNER JOIN "user" u
                    ON f.user_id = u.id
                    INNER JOIN "post" p
                    ON f.post_id = p.id
                    WHERE (u.age = 36) AND (DATE(f.time) = '2021-12-01')
                    AND (f.action = 'view') AND (p.topic = 'covid')
                    """, con)
```

```python
df
```

```python
df_grouped = df.groupby('city').agg(count=('topic', 'count'))
df_grouped.sort_values(by='count', ascending=True).reset_index().iloc[0]['city']
```

```python
top_liked_posts = pd.read_sql("""
                            SELECT post_id, COUNT(action), MAX(time)
                            FROM feed_action 
                            WHERE (action = 'like')
                            GROUP BY post_id
                            ORDER BY count DESC
                            LIMIT 10""", con)
```
