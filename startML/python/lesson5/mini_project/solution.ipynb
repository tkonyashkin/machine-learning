```python
unique_playstore = playstore.drop_duplicates(subset='App').reset_index(drop=True)
```

```python
new_columns = unique_playstore.columns.str.replace(' ', '_').str.lower()
unique_playstore.columns = new_columns
unique_playstore
```

```python
playstore = unique_playstore.copy()
```

```python
percent_free_apps = np.round((playstore['type'][unique_playstore['type'] == 'Free'].count() / unique_playstore['type'].count()), 2)
percent_free_apps
```

```python
playstore = unique_playstore.copy()
playstore[playstore['type'] != 'Free'].head()
```

```python
education_playstore = playstore[(playstore['category'] == 'EDUCATION') & (playstore['reviews'] > 1000)].reset_index(drop=True)
education_playstore.head()
```

```python
playstore['price'] = playstore['price'].str.replace('$', '').astype('float')
```

```python
playstore.head()
```

```python
playstore_pivot = playstore.pivot_table(index=['category', 'type'], values=['price', 'rating', 'reviews'], aggfunc=np.mean).round({'price': 2, 'rating': 1, 'reviews': 2})
playstore_pivot = playstore_pivot.rename(columns={'price': 'mean_price', 'rating': 'mean_rating', 'reviews': 'mean_reviews'})
playstore_pivot.to_csv('playstore_pivot.csv', sep=',')
playstore_pivot
```
